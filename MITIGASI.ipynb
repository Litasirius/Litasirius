```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

# Fungsi untuk plot mitigasi dengan input interaktif
def plot_mitigation(reduction_rate):
    years = np.arange(2020, 2051)
    emissions = np.linspace(100, 20 * reduction_rate, len(years))  # Pengurangan emisi berdasarkan input siswa
    plt.figure(figsize=(10, 5))
    plt.plot(years, emissions, label='Emisi CO2')
    plt.xlabel('Tahun')
    plt.ylabel('Emisi CO2 (juta ton)')
    plt.title('Simulasi Mitigasi: Pengurangan Emisi CO2')
    plt.legend()
    plt.grid(True)
    plt.show()

# Fungsi untuk plot adaptasi dengan input interaktif
def plot_adaptation(adaptation_rate):
    years = np.arange(2020, 2051)
    sea_level_rise = np.linspace(0, 100, len(years))  # Kenaikan permukaan laut tetap
    adaptation_efforts = np.linspace(0, 100 * adaptation_rate, len(years))  # Upaya adaptasi berdasarkan input siswa
    plt.figure(figsize=(10, 5))
    plt.plot(years, sea_level_rise, label='Kenaikan Permukaan Laut')
    plt.plot(years, adaptation_efforts, label='Upaya Adaptasi')
    plt.xlabel('Tahun')
    plt.ylabel('Kenaikan Permukaan Laut / Upaya Adaptasi')
    plt.title('Simulasi Adaptasi: Kenaikan Permukaan Laut dan Tindakan Adaptasi')
    plt.legend()
    plt.grid(True)
    plt.show()
# Interaksi dengan pengguna
interact(plot_mitigation, reduction_rate=(0.1, 1.0, 0.1))
interact(plot_adaptation, adaptation_rate=(0.1, 1.0, 0.1))
```
