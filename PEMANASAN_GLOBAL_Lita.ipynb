```python
# EMISI GAS RUMAH KACA #
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interactive, IntSlider, FloatSlider
import ipywidgets as widgets

def greenhouse_gas_simulation(initial_emission_rate, decay_rate, time):
    """
    Simulate greenhouse gas emissions over time.

    Parameters:
    - initial_emission_rate: Initial emission rate (tons/year)
    - decay_rate: Rate of emission decay (% per year)
    - time: Duration of simulation (years)
    """
    # Initialize variables
    time_values = np.arange(0, time + 1)
    emission_values = []

    # Calculate emissions for each year
    for t in time_values:
        emission = initial_emission_rate * np.exp(-decay_rate * t)
        emission_values.append(emission)

    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(time_values, emission_values, marker='o', linestyle='-')
    plt.xlabel('Time (years)')
    plt.ylabel('Emission Rate (tons/year)')
    plt.title('Greenhouse Gas Emission Simulation')
    plt.grid(True)
    plt.show()

    # Print total emissions
    total_emission = sum(emission_values)
    print(f'Total emissions over {time} years: {total_emission:.2f} tons')

# Create sliders for interactive parameters
initial_emission_slider = FloatSlider(min=100, max=10000, step=100, value=2000, description='Initial Emission Rate (tons/year)')
decay_rate_slider = FloatSlider(min=0.01, max=1, step=0.01, value=0.1, description='Decay Rate (% per year)')
time_slider = IntSlider(min=1, max=50, step=1, value=10, description='Duration (years)')

# Create interactive plot
interactive_plot = interactive(greenhouse_gas_simulation,
                               initial_emission_rate=initial_emission_slider,
                               decay_rate=decay_rate_slider,
                               time=time_slider)

# Display interactive plot
display(interactive_plot)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter model
years = np.arange(2020, 2100, 1)  # Tahun simulasi
initial_temp = 15.0  # Suhu rata-rata global awal (°C)
radiative_forcing_CO2 = 0.04  # W/m² per ppm
radiative_forcing_CH4 = 0.02  # W/m² per ppb
climate_sensitivity = 0.8  # °C per W/m²

# Konsentrasi gas rumah kaca (misalnya CO2 dan CH4)
co2_concentration = np.linspace(400, 700, len(years))  # ppm
ch4_concentration = np.linspace(1800, 2500, len(years))  # ppb

# Radiasi masuk dari matahari (W/m²)
solar_radiation = 340  # Konstanta radiasi matahari yang diterima per m²

# Penyerapan dan pelepasan radiasi
def calculate_radiative_forcing(co2, ch4):
    rf_co2 = radiative_forcing_CO2 * (co2 - 400)  # Perubahan dari level preindustri
    rf_ch4 = radiative_forcing_CH4 * (ch4 - 1800)
    return rf_co2 + rf_ch4

# Suhu rata-rata global
def calculate_temperature_change(forcing):
    return climate_sensitivity * forcing

# Array untuk menyimpan hasil
radiative_forcing = np.zeros(len(years))
temperature_change = np.zeros(len(years))
surface_temperature = np.zeros(len(years))

# Inisialisasi suhu permukaan awal
surface_temperature[0] = initial_temp

# Loop simulasi
for i in range(1, len(years)):
    rf = calculate_radiative_forcing(co2_concentration[i], ch4_concentration[i])
    radiative_forcing[i] = rf
    temp_change = calculate_temperature_change(rf)
    temperature_change[i] = temp_change
    surface_temperature[i] = initial_temp + temp_change

# Visualisasi hasil
fig, ax1 = plt.subplots(figsize=(12, 8))

# Konsentrasi CO2 dan CH4
ax1.set_xlabel('Tahun')
ax1.set_ylabel('Konsentrasi CO2 (ppm) dan CH4 (ppb)', color='tab:blue')
line1, = ax1.plot(years, co2_concentration, label='Konsentrasi CO2', color='tab:blue', linestyle='-', linewidth=2)
line2, = ax1.plot(years, ch4_concentration, label='Konsentrasi CH4', color='tab:cyan', linestyle='--', linewidth=2)
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.legend(loc='upper left')

# Suhu rata-rata global
ax2 = ax1.twinx()
ax2.set_ylabel('Suhu Rata-rata Global (°C)', color='tab:red')
line3, = ax2.plot(years, surface_temperature, label='Suhu Rata-rata Global', color='tab:red', linestyle='-.', linewidth=2)
ax2.tick_params(axis='y', labelcolor='tab:red')
ax2.legend(loc='upper right')

# Tambahan gaya visualisasi
plt.title('Simulasi Model Fisika Atmosfer: Interaksi Radiasi dan Gas Rumah Kaca', fontsize=16, fontweight='bold')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
fig.tight_layout()

# Menampilkan plot
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Parameter model
initial_CO2_emissions = 10  # GtCO2 per tahun, emisi CO2 tahunan awal
annual_emissions_reduction = 0.5  # GtCO2 per tahun, pengurangan emisi tanpa kebijakan tambahan
climate_sensitivity = 0.02  # °C per GtCO2, sensitivitas iklim (sederhana)
initial_temp = 14.0  # °C, suhu rata-rata global saat ini

# Kebijakan mitigasi
renewable_energy_increase = 0.5  # GtCO2 per tahun, pengurangan tambahan emisi dari energi terbarukan
carbon_tax_effect = 0.3  # GtCO2 per tahun, pengurangan tambahan emisi dari pajak karbon

# Waktu simulasi
years = 50
time = np.arange(0, years, 1)

# Array untuk menyimpan hasil
emissions_levels = np.zeros_like(time, dtype=float)
temperature_change = np.zeros_like(time, dtype=float)

# Inisialisasi nilai awal
emissions_levels[0] = initial_CO2_emissions
temperature_change[0] = initial_temp

# Loop simulasi
for t in range(1, years):
    # Emisi tanpa kebijakan tambahan
    emissions = emissions_levels[t-1] - annual_emissions_reduction

    # Tambahkan pengurangan emisi dari kebijakan mitigasi
    if t > 10:  # Asumsikan kebijakan diterapkan setelah 10 tahun
        emissions -= renewable_energy_increase
        emissions -= carbon_tax_effect

    emissions = max(emissions, 0)  # Emisi tidak bisa negatif
    emissions_levels[t] = emissions

    # Perhitungan perubahan suhu
    total_emissions_reduction = initial_CO2_emissions - emissions
    temp_increase = climate_sensitivity * total_emissions_reduction
    temperature_change[t] = initial_temp + temp_increase

# Plot hasil simulasi dengan plotly
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=time, y=emissions_levels,
    mode='lines+markers',
    name='Emisi CO2 (GtCO2 per tahun)',
    line=dict(color='blue', width=2),
    marker=dict(size=6)
))

fig.add_trace(go.Scatter(
    x=time, y=temperature_change,
    mode='lines+markers',
    name='Perubahan Suhu (°C)',
    line=dict(color='red', width=2),
    marker=dict(size=6),
    yaxis='y2'
))

# Menambahkan sumbu kedua untuk suhu
fig.update_layout(
    title='Simulasi Kebijakan Energi dan Pemanasan Global',
    xaxis=dict(title='Tahun'),
    yaxis=dict(title='Emisi CO2 (GtCO2 per tahun)', titlefont=dict(color='blue'), tickfont=dict(color='blue')),
    yaxis2=dict(title='Suhu Rata-rata Global (°C)', overlaying='y', side='right', titlefont=dict(color='red'), tickfont=dict(color='red')),
    legend=dict(x=0.05, y=0.95),
    template='plotly_white'
)

# Menampilkan plot interaktif
fig.show()
```

```python
# SIMULASI PEMANSAN GLOBAL #
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, FloatSlider
import ipywidgets as widgets

# Parameter simulasi
years = np.arange(2000, 2051)  # Rentang tahun
initial_carbon_level = 400  # Konsentrasi karbon awal (ppm)
renewable_energy_factor = 0.5  # Faktor penggunaan energi terbarukan
carbon_tax_rate = 20  # Tarif pajak karbon ($ per ton CO2)
carbon_tax_threshold = 450  # Ambang batas konsentrasi karbon (ppm)

# Fungsi simulasi pemanasan global
def simulate_global_warming(renewable_energy_factor, carbon_tax_rate):
    carbon_levels = [initial_carbon_level]
    for year in years[1:]:
        carbon_emission = (1 - renewable_energy_factor) * (year - 2000)  # Emisi karbon dari energi non-terbarukan
        if carbon_levels[-1] > carbon_tax_threshold:
            carbon_tax = carbon_tax_rate * (carbon_levels[-1] - carbon_tax_threshold)
        else:
            carbon_tax = 0
        net_emission = carbon_emission - carbon_tax
        carbon_levels.append(carbon_levels[-1] + net_emission)
    return carbon_levels

# Fungsi visualisasi dan animasi
def visualize_global_warming(renewable_energy_factor, carbon_tax_rate):
    carbon_levels = simulate_global_warming(renewable_energy_factor, carbon_tax_rate)
    plt.figure(figsize=(10, 6))
    plt.plot(years, carbon_levels, marker='o')
    plt.title('Simulasi Pemanasan Global')
    plt.xlabel('Tahun')
    plt.ylabel('Konsentrasi Karbon (ppm)')
    plt.grid(True)
    plt.show()

# Widget interaktif
interactive_plot = interactive(visualize_global_warming,
                               renewable_energy_factor=FloatSlider(min=0, max=1, step=0.1, value=0.5, description='Energi Terbarukan'),
                               carbon_tax_rate=FloatSlider(min=0, max=50, step=5, value=20, description='Pajak Karbon ($/ton CO2)'))
display(interactive_plot)
```

```python
# SIMULASI FEEDBACK LOOP DAN TIPPING POINT #
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, FloatSlider
import ipywidgets as widgets

# Parameter simulasi
years = np.arange(2000, 2051)  # Rentang tahun
initial_permafrost_coverage = 0.5  # Persentase tutupan permafrost awal
permafrost_melt_rate = 0.02  # Tingkat pencairan permafrost per tahun
methane_release_factor = 0.5  # Faktor pelepasan metana

# Fungsi simulasi feedback loop dan tipping points
def simulate_feedback_loop(permafrost_melt_rate, methane_release_factor):
    permafrost_coverage = [initial_permafrost_coverage]
    methane_levels = [0]
    for year in years[1:]:
        permafrost_melt = permafrost_melt_rate * permafrost_coverage[-1]
        methane_release = methane_release_factor * permafrost_melt
        permafrost_coverage.append(permafrost_coverage[-1] - permafrost_melt)
        methane_levels.append(methane_levels[-1] + methane_release)
    return permafrost_coverage, methane_levels

# Fungsi visualisasi dan animasi
def visualize_feedback_loop(permafrost_melt_rate, methane_release_factor):
    permafrost_coverage, methane_levels = simulate_feedback_loop(permafrost_melt_rate, methane_release_factor)

    plt.figure(figsize=(10, 6))
    plt.plot(years, permafrost_coverage, label='Tutupan Permafrost', marker='o')
    plt.plot(years, methane_levels, label='Konsentrasi Metana', marker='o')
    plt.title('Simulasi Feedback Loop dan Tipping Points')
    plt.xlabel('Tahun')
    plt.ylabel('Persentase Tutupan / Konsentrasi (%)')
    plt.grid(True)
    plt.legend()
    plt.show()

# Widget interaktif
interactive_plot = interactive(visualize_feedback_loop,
                               permafrost_melt_rate=FloatSlider(min=0.01, max=0.1, step=0.01, value=0.02, description='Tingkat Pencairan Permafrost'),
                               methane_release_factor=FloatSlider(min=0.1, max=1.0, step=0.1, value=0.5, description='Faktor Pelepasan Metana'))
display(interactive_plot)
```

```python
# SIMULASI INDIKATOR IKLIM UTAMA #
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, FloatSlider
import ipywidgets as widgets

# Parameter simulasi
years = np.arange(2000, 2051)  # Rentang tahun
initial_greenhouse_gas_concentration = 400  # Konsentrasi gas rumah kaca awal (ppm)
initial_global_temperature = 14  # Suhu rata-rata global awal (°C)
initial_polar_ice_coverage = 10  # Persentase tutupan es kutub awal (%)

# Fungsi simulasi indikator iklim utama
def simulate_climate_indicators(emission_factor, initial_temperature):
    greenhouse_gas_concentration = np.full_like(years, initial_greenhouse_gas_concentration)
    global_temperature = np.full_like(years, initial_global_temperature)
    polar_ice_coverage = np.full_like(years, initial_polar_ice_coverage)

    for i in range(1, len(years)):
        greenhouse_gas_concentration[i] = greenhouse_gas_concentration[i-1] * (1 + emission_factor)
        global_temperature[i] = global_temperature[i-1] + 0.02 * (greenhouse_gas_concentration[i] - initial_greenhouse_gas_concentration)
        polar_ice_coverage[i] = max(0, polar_ice_coverage[i-1] - 0.1 * (global_temperature[i] - initial_temperature))

    return greenhouse_gas_concentration, global_temperature, polar_ice_coverage

# Fungsi visualisasi dan animasi
def visualize_climate_indicators(emission_factor, initial_temperature):
    greenhouse_gas_concentration, global_temperature, polar_ice_coverage = simulate_climate_indicators(emission_factor, initial_temperature)

    plt.figure(figsize=(10, 6))
    plt.plot(years, greenhouse_gas_concentration, label='Konsentrasi Gas Rumah Kaca (ppm)', marker='o')
    plt.plot(years, global_temperature, label='Suhu Rata-rata Global (°C)', marker='o')
    plt.plot(years, polar_ice_coverage, label='Tingkat Es Kutub (%)', marker='o')
    plt.title('Simulasi Indikator Iklim Utama')
    plt.xlabel('Tahun')
    plt.grid(True)
    plt.legend()
    plt.show()

# Widget interaktif
interactive_plot = interactive(visualize_climate_indicators,
                               emission_factor=FloatSlider(min=0.01, max=0.05, step=0.01, value=0.03, description='Faktor Emisi'),
                               initial_temperature=FloatSlider(min=10, max=20, step=1, value=14, description='Suhu Awal (°C)'))
display(interactive_plot)
```

```python
# INTEGRASI DATA SATELIT DAN OBSERVASI SUHU #
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, FloatSlider
import ipywidgets as widgets

# Data historis dan satelit (simulasi)
years = np.arange(2000, 2021)
satellite_temperature = np.random.uniform(14.5, 15.5, size=len(years))  # Data suhu satelit (°C)
in_situ_temperature = satellite_temperature + np.random.normal(0, 0.1, size=len(years))  # Data suhu in situ dengan noise

# Fungsi simulasi
def simulate_temperature_change(factor_satelite, factor_in_situ):
    # Menggabungkan data satelit dan in-situ dengan faktor penyesuaian
    combined_temperature = (factor_satelite * satellite_temperature + factor_in_situ * in_situ_temperature) / (factor_satelite + factor_in_situ)
    return combined_temperature

# Fungsi visualisasi
def visualize_temperature_change(factor_satelite, factor_in_situ):
    combined_temperature = simulate_temperature_change(factor_satelite, factor_in_situ)

    # Plot hasil simulasi
    plt.figure(figsize=(10, 6))
    plt.plot(years, satellite_temperature, linestyle='--', color='blue', label='Satelit')
    plt.plot(years, in_situ_temperature, linestyle='--', color='green', label='In Situ')
    plt.plot(years, combined_temperature, marker='o', color='red', label='Gabungan')
    plt.title('Integrasi Data Satelit dan Observasi Suhu')
    plt.xlabel('Tahun')
    plt.ylabel('Suhu (°C)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Widget interaktif
factor_satelite_slider = FloatSlider(value=0.5, min=0.1, max=1.0, step=0.1, description='Faktor Satelit:')
factor_in_situ_slider = FloatSlider(value=0.5, min=0.1, max=1.0, step=0.1, description='Faktor In Situ:')

interactive_plot = interactive(visualize_temperature_change,
                               factor_satelite=factor_satelite_slider,
                               factor_in_situ=factor_in_situ_slider)

# Tampilan tabel percobaan interaktif
experiment_table = widgets.HTML(
    value="<table><tr><th>Percobaan</th><th>Faktor Satelit</th><th>Faktor In Situ</th></tr>"
)
display(experiment_table, interactive_plot)
```

```python
#SIMULASI CURAH HUJAN #
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, FloatSlider
import ipywidgets as widgets

# Parameter simulasi
years = np.arange(2000, 2051)  # Rentang tahun
base_rainfall = 100  # Curah hujan rata-rata dasar (mm)
extreme_rainfall_factor = 2  # Faktor peningkatan curah hujan ekstrem
extreme_rainfall_frequency = 0.1  # Frekuensi kejadian curah hujan ekstrem

# Fungsi simulasi curah hujan
def simulate_rainfall(intensity_factor, frequency_factor):
    base_rainfall_values = np.full_like(years, base_rainfall, dtype=np.float64)  # Ensure float type
    extreme_rainfall_values = np.zeros_like(years, dtype=np.float64)  # Ensure float type
    for i in range(len(years)):
        if np.random.rand() < extreme_rainfall_frequency * frequency_factor:
            extreme_rainfall_values[i] = base_rainfall * extreme_rainfall_factor
    total_rainfall_values = base_rainfall_values + extreme_rainfall_values
    total_rainfall_values *= intensity_factor
    extreme_rainfall_count = np.count_nonzero(extreme_rainfall_values)
    return total_rainfall_values, extreme_rainfall_count

# Fungsi visualisasi curah hujan
def visualize_rainfall(intensity_factor, frequency_factor):
    total_rainfall_values, extreme_rainfall_count = simulate_rainfall(intensity_factor, frequency_factor)
    plt.figure(figsize=(10, 6))
    plt.plot(years, total_rainfall_values, label='Total Curah Hujan', marker='o')
    plt.axhline(base_rainfall * intensity_factor, color='r', linestyle='--', label='Curah Hujan Rata-rata')
    plt.xlabel('Tahun')
    plt.ylabel('Curah Hujan (mm)')
    plt.title(f'Simulasi Curah Hujan (Intensitas: {intensity_factor}, Frekuensi: {frequency_factor})')
    plt.legend()
    plt.grid(True)
    plt.show()

# Widget interaktif
interactive_plot = interactive(visualize_rainfall,
                               intensity_factor=FloatSlider(min=0.5, max=2.0, step=0.1, value=1.0, description='Intensitas Hujan'),
                               frequency_factor=FloatSlider(min=0.01, max=0.5, step=0.01, value=0.1, description='Frekuensi Hujan Ekstrem'))
display(interactive_plot)
```

```python
# SIMULASI SIKLUS BIOGEOKIMIA : KARBON #
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, FloatSlider
import ipywidgets as widgets

# Parameter simulasi
years = np.arange(1990, 2021)  # Rentang tahun
carbon_emission = np.random.uniform(1, 3, size=len(years))  # Emisi karbon dalam gigaton CO2
carbon_absorption_forest = np.random.randint(10, 20, size=len(years))  # Penyerapan karbon oleh hutan
carbon_absorption_ocean = np.random.randint(5, 15, size=len(years))  # Penyerapan karbon oleh lautan

# Fungsi simulasi
def simulate_carbon_cycle(emission_factor_forest, emission_factor_ocean):
    total_carbon = np.zeros_like(carbon_emission, dtype=np.float64)
    total_carbon[0] = carbon_emission[0]  # Memulai dengan emisi awal

    for i in range(1, len(carbon_emission)):
        # Hitung total karbon
        total_carbon[i] = total_carbon[i-1] + carbon_emission[i] - \
                          emission_factor_forest * carbon_absorption_forest[i] - \
                          emission_factor_ocean * carbon_absorption_ocean[i]
    return total_carbon

# Fungsi visualisasi
def visualize_carbon_cycle(emission_factor_forest, emission_factor_ocean):
    total_carbon = simulate_carbon_cycle(emission_factor_forest, emission_factor_ocean)

    plt.figure(figsize=(10, 6))
    plt.plot(years, total_carbon, marker='o', label='Total Karbon (Gt CO2)')
    plt.plot(years, carbon_emission, linestyle='--', color='red', label='Emisi Karbon')
    plt.plot(years, emission_factor_forest * carbon_absorption_forest, linestyle='--', color='green', label='Penyerapan Karbon Hutan')
    plt.plot(years, emission_factor_ocean * carbon_absorption_ocean, linestyle='--', color='blue', label='Penyerapan Karbon Lautan')
    plt.title('Simulasi Siklus Biogeokimia: Karbon')
    plt.xlabel('Tahun')
    plt.ylabel('Karbon (Gt CO2)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Widget interaktif
emission_factor_forest_slider = FloatSlider(value=0.5, min=0.1, max=1.0, step=0.1, description='Faktor Emisi Hutan:')
emission_factor_ocean_slider = FloatSlider(value=0.3, min=0.1, max=1.0, step=0.1, description='Faktor Emisi Lautan:')

interactive_plot = interactive(visualize_carbon_cycle,
                               emission_factor_forest=emission_factor_forest_slider,
                               emission_factor_ocean=emission_factor_ocean_slider)

# Tampilan tabel percobaan interaktif
experiment_table = widgets.HTML(
    value="<table><tr><th>Percobaan</th><th>Faktor Emisi Hutan</th><th>Faktor Emisi Lautan</th></tr>"
)
display(experiment_table, interactive_plot)
```

```python
# Simulasi Pencairan Es di Kutub #
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interactive, IntSlider, FloatSlider
import ipywidgets as widgets
def ice_melting_simulation(surface_area, temperature, thickness, time):
    """
    Simulate ice melting at the poles.

    Parameters:
    - surface_area: Surface area of the ice (m^2)
    - temperature: Ambient temperature (°C)
    - thickness: Initial thickness of the ice (m)
    - time: Duration of simulation (years)
    """
    # Constants
    k = 2.1  # Thermal conductivity of ice (W/m°C)
    density = 917  # Density of ice (kg/m^3)
    specific_heat = 2100  # Specific heat of ice (J/kg°C)
    delta_t = 3600 * 24 * 365  # Time step (s)

    # Initialize variables
    timesteps = int(time * 365 * 24 * 60 * 60 / delta_t)
    ice_volume = surface_area * thickness
    melted_volume = 0
    temperature_data = [temperature]
    time_data = [0]

    for step in range(timesteps):
        temperature_change = (k * surface_area * (temperature - 0)) / (thickness * density * specific_heat) * delta_t
        temperature -= temperature_change
        melted_volume += (temperature_change * surface_area * delta_t) / (density * thickness)
        thickness -= (temperature_change * delta_t) / (density * surface_area)
        time_data.append((step + 1) * delta_t / (3600 * 24 * 365))
        temperature_data.append(temperature)

        if thickness <= 0:
            break

    plt.figure(figsize=(10, 6))
    plt.plot(time_data, temperature_data, label='Temperature (°C)')
    plt.xlabel('Time (years)')
    plt.ylabel('Temperature (°C)')
    plt.title('Ice Melting Simulation')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f'Total melted volume: {melted_volume:.2f} m^3')
    print(f'Remaining ice thickness: {thickness:.2f} m')
def ice_melting_simulation(surface_area, temperature, thickness, time):
    """
    Simulate ice melting at the poles.

    Parameters:
    - surface_area: Surface area of the ice (m^2)
    - temperature: Ambient temperature (°C)
    - thickness: Initial thickness of the ice (m)
    - time: Duration of simulation (years)
    """
    # Constants
    k = 2.1  # Thermal conductivity of ice (W/m°C)
    density = 917  # Density of ice (kg/m^3)
    specific_heat = 2100  # Specific heat of ice (J/kg°C)
    delta_t = 3600 * 24 * 365  # Time step (s)

    # Initialize variables
    timesteps = int(time * 365 * 24 * 60 * 60 / delta_t)
    ice_volume = surface_area * thickness
    melted_volume = 0
    temperature_data = [temperature]
    time_data = [0]

    for step in range(timesteps):
        temperature_change = (k * surface_area * (temperature - 0)) / (thickness * density * specific_heat) * delta_t
        temperature -= temperature_change
        melted_volume += (temperature_change * surface_area * delta_t) / (density * thickness)
        thickness -= (temperature_change * delta_t) / (density * surface_area)
        time_data.append((step + 1) * delta_t / (3600 * 24 * 365))
        temperature_data.append(temperature)

        if thickness <= 0:
            break

    plt.figure(figsize=(10, 6))
    plt.plot(time_data, temperature_data, label='Temperature (°C)')
    plt.xlabel('Time (years)')
    plt.ylabel('Temperature (°C)')
    plt.title('Ice Melting Simulation')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f'Total melted volume: {melted_volume:.2f} m^3')
    print(f'Remaining ice thickness: {thickness:.2f} m')
surface_area_slider = FloatSlider(min=1, max=100, step=1, value=50, description='Surface Area (m^2)')
temperature_slider = IntSlider(min=-50, max=10, step=1, value=-10, description='Temperature (°C)')
thickness_slider = FloatSlider(min=0.1, max=10, step=0.1, value=3, description='Ice Thickness (m)')
time_slider = FloatSlider(min=1, max=100, step=1, value=10, description='Duration (years)')

interactive_plot = interactive(ice_melting_simulation,
                               surface_area=surface_area_slider,
                               temperature=temperature_slider,
                               thickness=thickness_slider,
                               time=time_slider)

display(interactive_plot)
```

```python
# Simulasi penggunaan Energi terbarukan #
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interactive, FloatSlider
import ipywidgets as widgets

def renewable_energy_simulation(initial_energy, growth_rate, time):
    """
    Simulate renewable energy usage over time.

    Parameters:
    - initial_energy: Initial renewable energy usage (in TWh)
    - growth_rate: Annual growth rate of renewable energy usage (%)
    - time: Duration of simulation (years)
    """
    # Initialize variables
    time_values = np.arange(0, time + 1)
    energy_values = []

    # Calculate energy usage for each year
    for t in time_values:
        energy = initial_energy * ((1 + growth_rate / 100) ** t)
        energy_values.append(energy)

    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(time_values, energy_values, marker='o', linestyle='-')
    plt.xlabel('Time (years)')
    plt.ylabel('Renewable Energy Usage (TWh)')
    plt.title('Renewable Energy Usage Simulation')
    plt.grid(True)
    plt.show()

    # Print total energy usage
    total_energy = sum(energy_values)
    print(f'Total renewable energy usage over {time} years: {total_energy:.2f} TWh')

# Create sliders for interactive parameters
initial_energy_slider = FloatSlider(min=1, max=1000, step=10, value=100, description='Initial Energy (TWh)')
growth_rate_slider = FloatSlider(min=0, max=10, step=0.1, value=5, description='Growth Rate (%)')
time_slider = FloatSlider(min=1, max=50, step=1, value=10, description='Duration (years)')

# Create interactive plot
interactive_plot = interactive(renewable_energy_simulation,
                               initial_energy=initial_energy_slider,
                               growth_rate=growth_rate_slider,
                               time=time_slider)

# Display interactive plot
display(interactive_plot)
```

```python
# EFEK PADA POLA CUACA #
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, IntSlider, FloatSlider
import ipywidgets as widgets

# Data untuk simulasi
years = np.arange(2000, 2021)
base_temperature = np.random.uniform(14, 15, size=len(years))  # Suhu rata-rata global dalam derajat Celsius
el_nino_effect = np.random.uniform(0.1, 0.5, size=len(years))  # Pengaruh El Niño
la_nina_effect = np.random.uniform(-0.5, -0.1, size=len(years))  # Pengaruh La Niña

# Fungsi simulasi pola cuaca
def simulate_weather(greenhouse_gas_factor, el_nino_factor, la_nina_factor):
    adjusted_temperature = base_temperature + greenhouse_gas_factor
    el_nino_temperature = adjusted_temperature + (el_nino_effect * el_nino_factor)
    la_nina_temperature = adjusted_temperature + (la_nina_effect * la_nina_factor)
    return adjusted_temperature, el_nino_temperature, la_nina_temperature

# Fungsi visualisasi pola cuaca
def visualize_weather(greenhouse_gas_factor, el_nino_factor, la_nina_factor):
    adjusted_temperature, el_nino_temperature, la_nina_temperature = simulate_weather(greenhouse_gas_factor, el_nino_factor, la_nina_factor)

    plt.figure(figsize=(12, 6))
    plt.plot(years, adjusted_temperature, label='Suhu Rata-rata Global', color='blue')
    plt.plot(years, el_nino_temperature, label='Suhu dengan El Niño', color='red')
    plt.plot(years, la_nina_temperature, label='Suhu dengan La Niña', color='green')
    plt.xlabel('Tahun')
    plt.ylabel('Suhu (°C)')
    plt.title('Simulasi Efek pada Pola Cuaca')
    plt.legend()
    plt.grid(True)
    plt.show()

# Widget interaktif
greenhouse_gas_slider = FloatSlider(value=1.0, min=0.5, max=2.0, step=0.1, description='Faktor Gas Rumah Kaca:')
el_nino_slider = FloatSlider(value=1.0, min=0.5, max=2.0, step=0.1, description='Faktor El Niño:')
la_nina_slider = FloatSlider(value=1.0, min=0.5, max=2.0, step=0.1, description='Faktor La Niña:')
interactive_plot = interactive(visualize_weather, greenhouse_gas_factor=greenhouse_gas_slider, el_nino_factor=el_nino_slider, la_nina_factor=la_nina_slider)

# Tabel percobaan interaktif
experiment_table = widgets.HTML(
    value="<table><tr><th>Percobaan</th><th>Faktor Gas Rumah Kaca</th><th>Faktor El Niño</th><th>Faktor La Niña</th></tr>"
)

# Tampilan interaktif
display(experiment_table, interactive_plot)
```
