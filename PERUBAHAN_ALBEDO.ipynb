```python
# SIMULASI PERUBAHAN ALBEDO #
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, FloatSlider, widgets

# Parameter simulasi
years = np.arange(1990, 2021)  # Rentang tahun
albedo_forest = np.random.uniform(0.1, 0.2, size=len(years))  # Albedo hutan
albedo_agriculture = np.random.uniform(0.2, 0.3, size=len(years))  # Albedo pertanian

# Fungsi simulasi
def simulate_albedo_change(albedo_factor_forest, albedo_factor_agriculture):
    # Hitung total albedo
    total_albedo = (1 - albedo_factor_forest) * albedo_forest + (1 - albedo_factor_agriculture) * albedo_agriculture
    return total_albedo

# Fungsi visualisasi
def visualize_albedo_change(albedo_factor_forest, albedo_factor_agriculture):
    total_albedo = simulate_albedo_change(albedo_factor_forest, albedo_factor_agriculture)

    # Plot hasil simulasi
    plt.figure(figsize=(10, 6))
    plt.plot(years, total_albedo, marker='o', label='Total Albedo')
    plt.plot(years, albedo_forest, linestyle='--', color='green', label='Albedo Hutan')
    plt.plot(years, albedo_agriculture, linestyle='--', color='blue', label='Albedo Pertanian')
    plt.title('Simulasi Perubahan Albedo')
    plt.xlabel('Tahun')
    plt.ylabel('Albedo')
    plt.legend()
    plt.grid(True)
    plt.show()

# Widget interaktif
albedo_factor_forest_slider = FloatSlider(value=0.9, min=0.1, max=1.0, step=0.05, description='Faktor Albedo Hutan:')
albedo_factor_agriculture_slider = FloatSlider(value=0.9, min=0.1, max=1.0, step=0.05, description='Faktor Albedo Pertanian:')

interactive_plot = interactive(visualize_albedo_change,
                               albedo_factor_forest=albedo_factor_forest_slider,
                               albedo_factor_agriculture=albedo_factor_agriculture_slider)

# Tampilan tabel percobaan interaktif
experiment_table = widgets.HTML(
    value="<table><tr><th>Percobaan</th><th>Faktor Albedo Hutan</th><th>Faktor Albedo Pertanian</th></tr>"
)
display(experiment_table, interactive_plot)
```
