```python
# PEMODELAN PERUBAHAN EKSTREM CUACA #
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, IntSlider, FloatSlider
import ipywidgets as widgets

# Fungsi untuk mensimulasikan kejadian cuaca ekstrem
def simulate_extreme_weather(temp_increase, sea_level_rise):
    years = np.arange(2000, 2100)
    storms = temp_increase * np.log(years - 1999) + sea_level_rise * np.sqrt(years - 1999)
    droughts = temp_increase * (years - 1999) / 100
    floods = sea_level_rise * (years - 1999) / 100
    return years, storms, droughts, floods

# Fungsi untuk memvisualisasikan kejadian cuaca ekstrem
def visualize_extreme_weather(temp_increase, sea_level_rise):
    years, storms, droughts, floods = simulate_extreme_weather(temp_increase, sea_level_rise)

    plt.figure(figsize=(12, 6))

    plt.plot(years, storms, label='Badai Tropis', color='blue')
    plt.plot(years, droughts, label='Kekeringan', color='red')
    plt.plot(years, floods, label='Banjir', color='green')

    plt.xlabel('Tahun')
    plt.ylabel('Frekuensi Kejadian')
    plt.title('Pemodelan Grafik Perubahan Ekstrem Cuaca')
    plt.legend()
    plt.grid(True)
    plt.show()
    # Widget interaktif
temp_increase_slider = FloatSlider(value=2.0, min=0.5, max=5.0, step=0.1, description='Kenaikan Suhu (°C):')
sea_level_rise_slider = FloatSlider(value=0.5, min=0.1, max=2.0, step=0.1, description='Kenaikan Permukaan Laut (m):')
interactive_plot = interactive(visualize_extreme_weather, temp_increase=temp_increase_slider, sea_level_rise=sea_level_rise_slider)

# Tabel percobaan interaktif
experiment_table = widgets.HTML(
    value="""
    <table>
        <tr>
            <th>Percobaan</th>
            <th>Kenaikan Suhu (°C)</th>
            <th>Kenaikan Permukaan Laut (m)</th>
            <th>Frekuensi Badai Tropis</th>
            <th>Frekuensi Kekeringan</th>
            <th>Frekuensi Banjir</th>
        </tr>
        <tr>
            <td>1</td>
            <td>2.0</td>
            <td>0.5</td>
            <td>log(years - 1999) + sqrt(years - 1999)</td>
            <td>(years - 1999) / 100</td>
            <td>(years - 1999) / 100</td>
        </tr>
        <tr>
            <td>2</td>
            <td></td>
            <td></td>
            <td></td>
            <td></td>
            <td></td>
        </tr>
        <tr>
            <td>3</td>
            <td></td>
            <td></td>
            <td></td>
            <td></td>
            <td></td>
        </tr>
        <tr>
            <td>4</td>
            <td></td>
            <td></td>
            <td></td>
            <td></td>
            <td></td>
        </tr>
        <tr>
            <td>5</td>
            <td></td>
            <td></td>
            <td></td>
            <td></td>
            <td></td>
        </tr>
    </table>
    """
)# Tampilan interaktif
display(experiment_table, interactive_plot)
```
