```python
# SIMULASI MODEL FISIKA ATMOSFER #
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive, FloatSlider, Layout
import ipywidgets as widgets
from IPython.display import display
import pandas as pd

# Fungsi untuk simulasi interaktif
def simulate_atmosphere(greenhouse_gas_concentration):
    # Data untuk simulasi (misalnya, suhu)
    years = np.arange(1900, 2021)
    temperature = np.random.uniform(13, 15, size=len(years))  # Simulasi data

    # Menghitung dampak gas rumah kaca
    temperature += greenhouse_gas_concentration * 0.1  # Contoh sederhana

    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(years, temperature, marker='o', color='PURPLE')
    plt.title('Simulasi Model Fisika Atmosfer')
    plt.xlabel('Tahun')
    plt.ylabel('Suhu (Â°C)')
    plt.grid(True)
    plt.show()

    # Tabel hasil pengamatan
    observation_data = {'Tahun': years, 'Suhu (Â°C)': temperature}
    observation_df = pd.DataFrame(observation_data)
    display(observation_df)
# Widget interaktif untuk mengatur konsentrasi gas rumah kaca
greenhouse_gas_slider = FloatSlider(value=0.0, min=0.0, max=10.0, step=0.1,
                                    description='Konsentrasi Gas Rumah Kaca:',
                                    style={'description_width': 'initial'},
                                    layout=Layout(width='50%'))

# Fungsi untuk membuat tabel percobaan
def experiment_table(greenhouse_gas_concentration):
    print(f"Tabel Percobaan:")
    print(f"1. Ubah nilai konsentrasi gas rumah kaca menggunakan slider untuk melihat dampaknya pada suhu atmosfer.")

# Widget interaktif untuk tabel percobaan
interactive_experiment_table = interactive(experiment_table,
                                           greenhouse_gas_concentration=greenhouse_gas_slider)

# Menampilkan widget interaktif
display(interactive_experiment_table)

# Memanggil fungsi simulasi dengan nilai awal parameter
simulate_atmosphere(greenhouse_gas_slider.value)
```
